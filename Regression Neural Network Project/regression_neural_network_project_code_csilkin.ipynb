```python
# Import packages:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import torch
import imageio.v3 as iio
import skimage
from torch import nn
from torch.utils.data import TensorDataset, DataLoader
```

```python
file_path = "C:/Users/Student/Desktop/FSU -- Data Science Master's/2023 Spring/STA 5635 -- Applied Machine Learning"
image_file = file_path + "/" + "horse001b.png"

# Load and plot image:
image = iio.imread(uri=image_file)
plt.imshow(image)
```

```python
# Create lists of x and y coordinates and pixel values
x = []
y = []
I = []

## Load info into lists
for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        x.append(i)
        y.append(j)
        I.append(image[i,j])

# Convert x and y coordinate lists into Tensor objects
x = torch.tensor(x).float()
y = torch.tensor(y).float()

## Stack coordinates into one input matrix
coords = torch.stack((x,y),-1)

## Standardize coordinate values
sx=torch.std(coords, 0)
mx = torch.mean(coords,0)
coords = (coords-mx)/sx

# Convert list of pixel values into Tensor object
I = torch.tensor(I).float()

## Standardize
sy=torch.std(I)
my= torch.mean(I)
I = (I-my)/sy

# Create TensorDataSet
data=TensorDataset(coords,I)

# Create DataLoader object for training with minibatch size of 64
loader=DataLoader(data,batch_size=64,shuffle=True)
```

```python
# Neural Network with one hidden layer:
class Net(nn.Module):
    ## Instantiate
    def __init__(self):
        super().__init__()
        
        ### Output to hidden layer + ReLU
        self.fc1 = nn.Linear(2,128)
        self.relu1 = nn.ReLU()
        ### Final output layer + ReLU
        self.fc2 = nn.Linear(128,1)
        self.relu2 = nn.ReLU()
    
    ## Feed-forward path
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu1(out)
        out = self.fc2(out)
        out = self.relu2(out)
        return out
```

```python
# Neural Network with two hidden layers:
class Net_2(nn.Module):
    ## Instantiate
    def __init__(self):
        super().__init__()
        
        ### Output to first hidden layer + ReLU
        self.fc1 = nn.Linear(2,32)
        self.relu1 = nn.ReLU()
        ### Output to second hidden layer + ReLU
        self.fc2 = nn.Linear(32,128)
        self.relu2 = nn.ReLU()
        ### Final output layer + ReLU
        self.fc3 = nn.Linear(128,1)
        self.relu3 = nn.ReLU()
    
    ## Feed-forward path
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu1(out)
        out = self.fc2(out)
        out = self.relu2(out)
        out = self.fc3(out)
        out = self.relu3(out)
        return out
```

```python
# Neural Network with three hidden layers:
class Net_3(nn.Module):
    ## Instantiate
    def __init__(self):
        super().__init__()
        
        ### Output to first hidden layer + ReLU
        self.fc1 = nn.Linear(2,32)
        self.relu1 = nn.ReLU()
        ### Output to second hidden layer + ReLU
        self.fc2 = nn.Linear(32,64)
        self.relu2 = nn.ReLU()
        ### Output to third hidden layer + ReLU
        self.fc3 = nn.Linear(64,128)
        self.relu3 = nn.ReLU()
        ### Final output layer + ReLU
        self.fc4 = nn.Linear(128,1)
        self.relu4 = nn.ReLU()
    
   	    ## Feed-forward path
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu1(out)
        out = self.fc2(out)
        out = self.relu2(out)
        out = self.fc3(out)
        out = self.relu3(out)
        out = self.fc4(out)
        out = self.relu4(out)
        return out
```

```python
# Neural Network with four hidden layers:
class Net_4(nn.Module):
    ## Instantiate
    def __init__(self):
        super().__init__()
        
        ### Output to first hidden layer + ReLU
        self.fc1 = nn.Linear(2,32)
        self.relu1 = nn.ReLU()
        ### Output to second hidden layer + ReLU
        self.fc2 = nn.Linear(32,64)
        self.relu2 = nn.ReLU()
        ### Output to third hidden layer + ReLU
        self.fc3 = nn.Linear(64,128)
        self.relu3 = nn.ReLU()
        ### Output to fourth hidden layer + ReLU
        self.fc4 = nn.Linear(128,128)
        self.relu4 = nn.ReLU()
        ### Final output layer + ReLU
        self.fc5 = nn.Linear(128,1)
        self.relu5 = nn.ReLU()
    
    ## Feed-forward path
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu1(out)
        out = self.fc2(out)
        out = self.relu2(out)
        out = self.fc3(out)
        out = self.relu3(out)
        out = self.fc4(out)
        out = self.relu4(out)
        out = self.fc5(out)
        out = self.relu5(out)
        return out
```

```python
"""
NEURAL NETWORK IMPLEMENTATION
"""
# Create instance of Neural Network (either Net(), Net_2(), Net_3(), or Net_4())
net=Net()

# List of losses
losses = []

# Iterate over epochs
for e in range(300):
    
    ## Initialize lambda
    lam = 0.003
    ### Divide lambda by 2 every 100 epochs
    if e >= 100:
        lam = 0.003/2
    elif e >= 200:
        lam = (0.003/2)/2
    
    ## Create optimizer
    optimizer = torch.optim.SGD(net.parameters(), lr=lam)
    ## Initialize loss
    sumloss=0
    
    ## Iterate through minibatches
    for xi,yi in loader:
        
        ### Neural network output w/ minibatch input
        out=net(xi).squeeze()
        
        ### Loss value for minibatch
        loss=torch.mean((out-yi)**2)
        
        ### Backpropagation
        
        #### Zero-out gradient
        optimizer.zero_grad()
        #### Back-propagate through neural network
        loss.backward()
        #### Update gradient
        optimizer.step()
        
        ### Sum loss terms from minibatches into final loss
        sumloss+=loss.item()*xi.shape[0]
        
    ## Append loss item to list of losses
    losses.append(sumloss)

# Plot losses vs. Iteration Number:
plt.plot(losses, color = "red")
plt.title("Training Loss vs Epoch Number -- Four Hidden Layers NN")
plt.xlabel("Epoch Number")
plt.ylabel("Training Loss")

# Load final output from neural network
op_test = net(coords)
# Convert to 84x128 image
im_test = op_test.detach().numpy().reshape(84,128)
# Plot image with pixel values derived from neural network
plt.imshow(im_test)
```
